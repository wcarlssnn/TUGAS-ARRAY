```python
nilai = []  # array kosong untuk menyimpan nilai

print("Masukkan 10 nilai mahasiswa:")
for i in range(10):
    n = float(input(f"Nilai ke-{i+1}: "))
    nilai.append(n)

# Tampilkan semua nilai
print("\nData nilai:", nilai)

# 2. Nilai tertinggi & terendah
nilai_tertinggi = max(nilai)
nilai_terendah = min(nilai)

print("Nilai tertinggi:", nilai_tertinggi)
print("Nilai terendah:", nilai_terendah)

# 3. Rata-rata
rata_rata = sum(nilai) / len(nilai)
print("Rata-rata:", rata_rata)

# 4. Hitung jumlah mahasiswa lulus (>= 60)
jumlah_lulus = 0
for n in nilai:
    if n >= 60:
        jumlah_lulus += 1

print("Jumlah mahasiswa lulus:", jumlah_lulus)

# 5 & 6. Grafik
import matplotlib.pyplot as plt

# Grafik nilai tertinggi & terendah
plt.figure()
plt.bar(["Tertinggi", "Terendah"], [nilai_tertinggi, nilai_terendah])
plt.title("Perbandingan Nilai Tertinggi & Terendah")
plt.show()

# Grafik kelulusan
tidak_lulus = len(nilai) - jumlah_lulus

plt.figure()
plt.pie([jumlah_lulus, tidak_lulus], labels=["Lulus", "Tidak Lulus"], autopct="%1.1f%%")
plt.title("Persentase Kelulusan")
plt.show()
```
